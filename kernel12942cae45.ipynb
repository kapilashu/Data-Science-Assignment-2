```python
def data_cleaning(train):
    train['dependency']=np.sqrt(train['SQBdependency'])
    train['rez_esc']=train['rez_esc'].fillna(0)
    train['v18q1']=train['v18q1'].fillna(0)
    train['v2a1']=train['v2a1'].fillna(0)
    
    conditions = [
    (str(train['edjefe'])=='no') & (str(train['edjefa'])=='no'), #both no
    (str(train['edjefe'])=='yes') & (str(train['edjefa'])=='no'), # yes and no
    (str(train['edjefe'])=='no') & (str(train['edjefa'])=='yes'), #no and yes 
    (str(train['edjefe'])!='no') & (str(train['edjefe'])!='yes') & (str(train['edjefa'])=='no'), # number and no
    (str(train['edjefe'])=='no') & (str(train['edjefa'])!='no') # no and number
    ]
    choices = [0, 1, 1, train['edjefe'], train['edjefa']]
    train['edjefx']=np.select(conditions, choices)
    train['edjefx']=train['edjefx'].astype(int)
    ##train.drop(['edjefe', 'edjefa'], axis=1, inplace=True)
    
    meaneduc_nan=train[train['meaneduc'].isnull()][['Id','idhogar','escolari']]
    me=meaneduc_nan.groupby('idhogar')['escolari'].mean().reset_index()
    for row in meaneduc_nan.iterrows():
        idx=row[0]
        idhogar=row[1]['idhogar']
        m=me[me['idhogar']==idhogar]['escolari'].tolist()[0]
        train.at[idx, 'meaneduc']=m
        train.at[idx, 'SQBmeaned']=m*m
        
    return train
```

# Modeling

## Light Gradient Boosting Machine

### Baseline Model

```python
features = [c for c in train.columns if c not in ['Id', 'Target']]
```

```python
target = train['Target'].values
target_index = {1:0, 2:1, 3:2, 4:3}
target = np.array([target_index[c] for c in target])
```

```python
def label_encoding(col):
    le = LabelEncoder()
    le.fit(list(train[col].values) + list(test[col].values))
    train[col] = le.transform(train[col].astype(str))
    test[col] = le.transform(test[col].astype(str))

num_cols = train._get_numeric_data().columns
cat_cols = list(set(features) - set(num_cols))
for col in cat_cols:
    label_encoding(col)
```

### Prepare Training and Validation Sets


```python
X_train, X_valid, y_train, y_valid = train_test_split(train[features].values, target, test_size=0.2, random_state=1)
lgb_train = lgb.Dataset(X_train, y_train)
lgb_valid = lgb.Dataset(X_valid, y_valid)
```

```python
params = {'boosting_type': 'gbdt', 'objective': 'multiclass', 'metric': 'multi_logloss',
          'num_class': 4, 'max_depth': 44, 'num_leaves': 36, 'learning_rate': 0.01,
          'feature_fraction': 0.8, 'bagging_fraction': 0.8, 'bagging_freq': 5,
          'lambda_l2': 1.0, 'verbose': -1, 'num_threads': -1 }

model = lgb.train(params, lgb_train, num_boost_round=100,  valid_sets=[lgb_train, lgb_valid], 
                  early_stopping_rounds=2000, verbose_eval=100)
```

### Predict and Submit


```python
test_data=data_cleaning(test)
features1 = [c for c in train.columns if c not in ['Id', 'Target']]

## predictions
preds = model.predict(test_data[features1], num_iteration=model.best_iteration)

## revert back to original classes
reverse_index = {}
for k,v in target_index.items():
    reverse_index[v] = k
preds = [reverse_index[np.argmax(p)] for p in preds]

## submission
subs = pd.DataFrame()
subs['Id'] = test_data['Id']
subs['Target'] = preds
subs.to_csv('submission_LGBM.csv', index=False)
subs.head()
```
